```python
fre = {} 
#df['average_rating']=df['average_rating'].astype('float64')
for item in df['Overall']: 
        if (item in fre): 
            fre[item] += 1
        else: 
            fre[item] = 1
key=[x for x in fre.keys()]
key
```

```python

value=[fre[x] for x in fre.keys()]
value
```

```python

import matplotlib.pyplot as plt
plt.pie(value,labels=key,startangle=90)
plt.show()
```

```python
plt.bar(key,value)
plt.show()
```

```python
plt.scatter(value, key)
plt.title('Overall of all players')
plt.xlabel('Count')
plt.ylabel('Overall')
freq = {} 

for item in df['Overall']: 
        if (item in freq): 
            freq[item] += 1
        else: 
            freq[item] = plt.show()
```
