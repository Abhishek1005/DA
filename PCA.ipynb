```
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from numpy import linalg as la
```

```

iris = datasets.load_iris()

X = iris.data
y = iris.target
target_names = iris.target_names
```

```
plt.figure()
colors = ['blue', 'red', 'green']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X[y == i, 0], X[y == i, 1], color=color, alpha=.9, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
```

```
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)
```

```
plt.figure()
colors = ['blue', 'red', 'green']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.9, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA')
```

```

X_rev = pca.inverse_transform(X_r)
X_revv = X_r.dot(pca.components_) + pca.mean_
```

```
loss = ((X - X_revv) ** 2).mean()
loss
```
