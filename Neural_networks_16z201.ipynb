```
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
```

```
from sklearn.model_selection import train_test_split
```

```
a=np.array([[1,0,1]])
b=np.array([[1,0]])
print(a)
print(b)
```

```
model=Sequential()
h1=model.add(Dense(2,activation="sigmoid",input_dim=3))
h2=model.add(Dense(3,activation="sigmoid"))

o = model.add(Dense(2,activation="sigmoid"))
```

```
wt_l1=np.array([[0.2,-0.3],[0.4,0.1],[-0.5,0.2]])
bias1=np.array([-0.4,0.2])
l=[]
l.append(wt_l1)
l.append(bias1)
model.layers[0].set_weights(l)
model.layers[0].get_weights()
```

```
wt_l2=np.array([[-0.3,0.2,0.6],[-0.2,-0.1,-0.2]])
bias2=np.array([0.1,0.1,0.2])
l=[]
l.append(wt_l2)
l.append(bias2)
model.layers[1].set_weights(l)
model.layers[1].get_weights()
```

```
wt_l3=np.array([[0.3,0.2],[-0.5,0.5],[-0.7,0.2]])
bias3=np.array([0.1,0.2])
l=[]
l.append(wt_l3)
l.append(bias3)
model.layers[2].set_weights(l)
model.layers[2].get_weights()
```

```
from keras.optimizers import Adam
model.compile(loss="mean_squared_error",optimizer=Adam(lr=0.9),metrics=['accuracy'])
```

```
model.summary()
```

```
model.fit(a,b,batch_size=1,epochs=1)
```
